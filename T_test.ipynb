```python
import numpy as np
from scipy import stats
```

```python
sample = [172, 174, 168, 169, 171, 173, 175, 170, 169, 172]
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)
mean_poplulation = 170
```

```python
T_stats = (sample_mean - mean_poplulation) / (sample_std / np.sqrt(n))
T_stats
```

```python
p_values = 2 * (1  - stats.t.cdf(abs(T_stats), df = n - 1))
alpha = 0.05
```

```python
if p_values < alpha:
    print("We will reject the Hypothesis")
else:
    print("We will accept the Hypothesis")
```
