```python
# Reverse a NumPy array without using a loop, using a slicing trick (don't use the simple arr[::-1] — try a different slicing approach to achieve the same result).
a = np.array([1,2,3,4])
np.flip(a)
```

```python
# Create two arrays - one with shape (3,1) and another with shape (1,4). Add them together using broadcasting, and observe what shape the result comes out to be.
A = np.array([1,2,3]).reshape(3,1)
B = np.array([5,6,7,8]).reshape(1,4)
A + B
```

```python
#Take an array, and compute its square root, square, and cube all together — then stack the three results as columns into one array (using np.column_stack).A
arr = np.array([1,4,9,16])
a = np.sqrt(arr)
b = np.square(arr)
c = arr ** 3
np.column_stack([a, b, c])
```

```python
#Create an array of Celsius temperatures, then convert it to Fahrenheit using the formula F = C*9/5 + 32, all in a single line (no loop needed).
C = np.array([20,35,40,8])
F = C * (9/5) + 32

print(F)

```
