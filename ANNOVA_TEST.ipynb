```python
import numpy as np
import seaborn as sns
import pandas as pd
from scipy.stats import f_oneway
```

```python
df = sns.load_dataset('titanic')
df
```

```python
df = df[['age', 'pclass']].dropna()
```

```python
df['pclass'].unique()
```

```python
class_1  = df[df['pclass'] == 1]['age']
class_2  = df[df['pclass'] == 2]['age']
class_3  = df[df['pclass'] == 3]['age']
class_1
```

```python
f_stats, p_value = f_oneway(class_1, class_2, class_3)
```

```python
f_stats
```

```python
p_value
```

```python
alpha = 0.05
```

```python
if p_value < alpha:
    print("Reject the null hypothesis there is a significant diff btw atleast one pclass")
else:
    print("There is a significant diff")
```
